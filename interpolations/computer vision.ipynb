```python
def resize_NN(img, shape):
  
    H, W = img.shape
    H_n, W_n = shape
    A_p = np.zeros(shape)
  
    for i in range(H_n):
        for j in range(W_n):
            x, y = (2 * i + 1) / (2 * H_n), (2 * j + 1) / (2 * W_n)
            A_p[i, j] = img[int(x * H),int(y * W)]

    return A_p
```

```python
def resize_by_factor(img, resize_factor):
  
    H, W = img.shape
    H_n, W_n = (H * resize_factor, W * resize_factor)
    A_p = np.zeros((H_n, W_n))
  
    for i in range(H_n):
        for j in range(W_n):
            x, y = (i / resize_factor, j / resize_factor)
#             x, y = (2 * i + 1) / (2 * H_n), (2 * j + 1) / (2 * W_n)
            #rint('x*H', x*H)
            A_p[i, j] = img[int(x), int(y)]

    return A_p
```

```python
# get two dimensi
def resize_NN_channels(img, shape):
  
    H, W, c = img.shape
    H_n, W_n = shape
    A_p = np.zeros((H_n, W_n, c))
  
    for i in range(H_n):
        for j in range(W_n):
            for k in range(c):
                x, y = (2 * i + 1) / (2 * H_n), (2 * j + 1) / (2 * W_n)
                A_p[i, j, k] = img[int(x * H),int(y * W), k]
                

    return A_p
```

### bilinear interpolation

```python
def bilinear_interpolation(x, y, k, img):
    
    x2 = int(ceil(x))
    x1 = int(floor(x))
    y2 = int(ceil(y))
    y1 = int(floor(y))
       
    if x2 == img.shape[0] or x1 == img.shape[0] or y2 == img.shape[1] or y1 == img.shape[1]:
        return 0
    
    ratio_x = x2 - x1
    ratio_y = y2 - y1
    
    interpolate_x_y1 = ((x2 - x) / ratio_x) * img[x1, y1, k] + ((x - x1) / ratio_x) * img[x2, y1, k]
    interpolate_x_y2 = ((x2 - x) / ratio_x) * img[x1, y2, k] + ((x - x1) / ratio_x) * img[x2, y2, k]
    interpolate_y = ((y2 - y) / ratio_y) * interpolate_x_y1 + ((y - y1) / ratio_y) * interpolate_x_y2
    
    return interpolate_y
```

```python
def bilinear_resize(filename, shape):
    original_image = cv2.imread(filename, -1)
    rows, cols, channels = original_image.shape
    print(rows, cols)
    new_image = np.zeros((shape[0], shape[1], channels))
    print(new_image.shape)

    for i in range(0, new_image.shape[0]):
        for j in range (0, new_image.shape[1]):
            for k in range (new_image.shape[2]):
				#### (2 * j + 1) / (2 * new_image.shape[1]), (2 * i + 1) / (2 * new_image.shape[0])
                new_image[i][j][k] = bilinear_interpolation( rows * (2 * i + 1) / (2 * new_image.shape[0]), cols * (2 * j + 1) / (2 * new_image.shape[1]), k, original_image)

    cv2.imwrite('khatiko_resized_by_500_300.jpg', new_image)

    
bilinear_resize('khatiko.jpg', (500, 300))
```

```python
def interpolate(first_value, second_value, ratio):
    # used linear weighted sum
    return first_value * (1 - ratio) + second_value * ratio

def bilinear_interpolation_new(x, y, c, img):
    
    x_rounded_up = int(ceil(x))
    x_rounded_down = int(floor(x))
    y_rounded_up = int(ceil(y))
    y_rounded_down = int(floor(y))
    
    if x_rounded_up == img.shape[0] or x_rounded_down == img.shape[0] or y_rounded_up == img.shape[1] or y_rounded_down == img.shape[1]:
        return 0

    ratio_x = x - x_rounded_down
    ratio_y = y - y_rounded_down

    interpolate_x1 = interpolate(img[x_rounded_down, y_rounded_down, c], img[x_rounded_down, y_rounded_up, c], ratio_x)
    interpolate_x2 = interpolate(img[x_rounded_up, y_rounded_down, c], img[x_rounded_up, y_rounded_up, c], ratio_x)
    interpolate_y = interpolate(interpolate_x1, interpolate_x2, ratio_y)
    
    return interpolate_y
```

```python
def bilinear_resize_by_factor(filename, resizing_factor):
    original_image = cv2.imread(filename, -1)
    rows, cols, channels = original_image.shape
    print(rows, cols)
    new_image = np.zeros(shape =(rows*resizing_factor, cols*resizing_factor, channels))
    print(new_image.shape)

    for i in range(0, new_image.shape[0]):
        for j in range (0, new_image.shape[1]):
            for k in range (new_image.shape[2]):
				#### (2 * j + 1) / (2 * new_image.shape[1]), (2 * i + 1) / (2 * new_image.shape[0])
                new_image[i][j][k] = bilinear_interpolation( cols * (2 * j + 1) / (2 * new_image.shape[1]),  rows * (2 * i + 1) / (2 * new_image.shape[0]), k, original_image)

    cv2.imwrite('khatiko_resized_by_2_new.jpg', new_image)

    
bilinear_resize_by_factor('khatiko.jpg', 2)
```

### rotate nearest neighbor

```python
def rotate(img, alpha, b = [0, 0]):
    A_alpha = np.array([[np.cos(alpha), -np.sin(alpha)],
                     [np.sin(alpha), np.cos(alpha)]])
  
    A_alpha_inv = np.linalg.inv(A_alpha)
  
    H, W = img.shape
    A_p = np.zeros_like(img)
  
    for i in range(H):
        for j in range(W):
            x, y = np.dot(A_alpha_inv, [(2 * i + 1) / (2 * H) - b[0], (2 * j + 1) / (2 * W) - b[1]])
            x, y = x + b[0], y + b[1]
            if 0 <= x < 1 and 0 <= y < 1:
                A_p[i, j] = img[int(x*H), int(y*W)]
  
    return A_p
```

```python
rotation_angle = np.deg2rad(20)
grayed_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rotated = rotate(grayed_image, rotation_angle, [0, 0])
plt.imshow(img_rotated)
```
